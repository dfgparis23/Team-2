```python
import pandas as pd
import numpy as np 
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python
invoice = pd.read_excel('../shared_data_read_only/Invoice and Opportunities.xlsx')
cpi = pd.read_excel('../shared_data_read_only/consumer_price_index.xlsx')
cpi = cpi.iloc[:,0:3]
cpi.head()
```

```python
invoice['Month'] = invoice['Date'].dt.to_period('M')
cpi['Month'] = cpi['Date'].dt.to_period('M')

# Group by the "Month" column and sum the "Amount"
monthly_invoice = invoice.groupby('Month').agg({"Amount": "sum"}).reset_index()
monthly_invoice.head()

merged_df = monthly_invoice.merge(cpi, on = "Month", how="outer")
merged_df.head()
```

```python
#merged_df['Month'] = merged_df['Month'].dt.strftime('%b %Y')

# Create the figure and axis objects
fig, ax1 = plt.subplots()

# Plot Amount_standard on the primary y-axis
ax1.set_xlabel('Month')
ax1.set_ylabel('Amount Standard', color='tab:blue')
ax1.plot(merged_df['Month'], merged_df['Amount_standard'], color='tab:blue', marker='o', label='Amount Standard')
ax1.tick_params(axis='y', labelcolor='tab:blue')

# Create a secondary y-axis
ax2 = ax1.twinx()
ax2.set_ylabel('Inflation Rate Standard', color='tab:red')
ax2.plot(merged_df['Month'], merged_df['Inflation_rate_standard'], color='tab:red', marker='x', label='Inflation Rate Standard')
ax2.tick_params(axis='y', labelcolor='tab:red')

# Add a title and legend
plt.title('Monthly Amount and Inflation Rate (Standard)')
fig.legend(loc="upper right")

# Rotating x-axis labels for better readability
plt.xticks(rotation=45)

# Display the plot
plt.tight_layout()
plt.show()
```

```python
# x min max standarnization  x = x - xmin/ (xmax-xmin)
merged_df['Amount_standard'] = ( merged_df['Amount'] - merged_df['Amount'].min())/ (merged_df['Amount'].max() - merged_df['Amount'].min())
merged_df
merged_df['Inflation_rate_standard'] = ( merged_df['Inflation_rate '] - merged_df['Inflation_rate '].min())/ (merged_df['Inflation_rate '].max() - merged_df['Inflation_rate '].min())
merged_df.head()
```

```python
#Kolmogorov-Smirnov test

from scipy.stats import kstest

#perform Kolmogorov-Smirnov test
kstest(merged_df['Amount'], 'norm')

from scipy.stats import ks_2samp

#perform Kolmogorov-Smirnov test
ks_2samp(merged_df['Amount_standard'], merged_df['Inflation_rate_standard'])

# We have sufficient evidence to say that the two sample datasets do come from the same distribution.
```

```python
# Create a figure and primary y-axis
fig, ax1 = plt.subplots(figsize=(10, 6))
ax1.set_xlabel('Month')
ax1.set_ylabel('Amount', color='blue')
ax1.plot(merged_df['Month'], merged_df['Amount_standard'], 'bo-', label='Amount')
ax1.tick_params(axis='y', labelcolor='blue')

# Create a secondary y-axis
ax2 = ax1.twinx()
ax2.set_ylabel('Inflation Rate', color='red')
ax2.plot(merged_df['Month'], merged_df['Inflation_rate_standard'], 'rx-', label='Inflation Rate')
ax2.tick_params(axis='y', labelcolor='red')

# Add a title and legend
plt.title('Monthly Amount and Inflation Rate')
fig.legend(loc="upper right")

# Rotating x-axis labels for better readability
plt.xticks(rotation=45)

# Display the plot
plt.tight_layout()
plt.show()
```

```python
items = pd.read_excel('../shared_data_read_only/Items and Opportunities.xlsx')
non_stock_item = pd.read_excel('../shared_data_read_only/Non-Stock Items.xlsx')
non_stock_item .head()

merged_items = items.merge(non_stock_item, on = "Inventory ID", how="left")
merged_items.head()
```

```python
grouped = merged_items.groupby('Inventory ID').size()

# Sort in descending order and select the top 20
top_20_descending = grouped.sort_values(ascending=False).head(20)

print(top_20_descending)
```

```python
inventory_ext_price = merged_items.groupby('Inventory ID').agg({'Ext. Price': 'sum'}).reset_index()
Tax_Category_ext_price = merged_items.groupby('Tax Category_x').agg({'Ext. Price': 'sum'}).reset_index()
Tax_Category_ext_price
```

```python
data = Tax_Category_ext_price
plt.figure(figsize=(10, 8))
bars = plt.bar(data['Tax Category_x'], data['Ext. Price'], color='skyblue', alpha=0.7)

plt.xlabel('Tax Category')
plt.ylabel('Ext. Price')
plt.title('Ext. Price by Tax Category')

plt.xticks(rotation=45)

for bar in bars:
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), f'{bar.get_height():.2f}', ha='center', va='bottom')

plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.grid(axis='y', linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```

```python
Opportunities = pd.read_excel('data/Opportunities.xlsx')
contacts = pd.read_excel('data/Contacts.xlsx')
merged_opportunity = pd.merge(Opportunities, contacts, right_on='ContactID', left_on='Contact ID', how='left')
```

```python
selected_columns = merged_opportunity[["Class_y", "Contact ID"]].dropna()

top20 = selected_columns.groupby('Class_y').size().sort_values(ascending=False).head(20)

plt.figure(figsize=(10, 6))  
top20.plot(kind='bar', color='skyblue')

plt.xlabel('Class')
plt.ylabel('Count')
plt.title('Top 20 Classes')

plt.xticks(range(len(top20)), top20.index, rotation=90)

plt.tight_layout()
plt.show()
```

```python

estimated_amount = merged_opportunity.groupby('Class_y')['Estimated amount'].sum().sort_values(ascending=False)
plt.figure(figsize=(10, 6))  
estimated_amount.plot(kind='bar', color='skyblue')

plt.xlabel('Class')
plt.ylabel('Count')
plt.title('Estimated amount by Classes')

plt.xticks(range(len(estimated_amount)), estimated_amount.index, rotation=90)

plt.tight_layout()
plt.show()
```
